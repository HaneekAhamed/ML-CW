```python
import pandas as pnd#IMPORT PANDA LIBRARY RENAME AS PND
#READ THE FILE (SKIP FIRST 36 ROW ,SEPARATED BY COLON,USE PYTHON ENGINE TO PASSING THE FILE, GIVE NAMES TO THE COLUMN IN DATA)
Topic=pnd.read_csv('spam_base.names', skiprows=36, sep=':\s+', engine='python', names=['attr',''])
Topic=Topic['attr']#ASSIGN NEW VARIABLE
Topic= list(Topic) # CONVERT VARIABLE TO LIST
Topic.append('spam_data')#APPEND SPAM_DATA
Topic
```

```python
#TELL PANDAS THERE IS NO HEADER
Data=pnd.read_csv('spam_base.data', names=Topic, header=None)
```

```python
#assinging the resulting data to new variable x
x = Data.drop(columns=['spam_data'])
x
```

```python
#assinging the resulting series  to new variable y
y = Data['spam_data']
y
```

```python
#this function used to split a dataset
from sklearn.model_selection import train_test_split
#30% testing data and x,y set to training and testing set of features
x_trains, x_tests, y_trains, y_tests = train_test_split(x, y, test_size=0.3,random_state=35)
```

```python
#used to standardize features by removing the scaling to unit variance
from sklearn.preprocessing import StandardScaler
standar_scaler = StandardScaler()
standar_scaler.fit(x)#fit the scaler to entire dataset
scaled_data = standar_scaler.transform(x_trains)#transform the training set feature
X_test_scaled = standar_scaler.transform(x_tests)#transform the testing set feature
```

```python
#the principal component analysis used for reducing the dimensionality
from sklearn.decomposition import PCA

PCA = PCA()
x_PCA = PCA.fit_transform(scaled_data)#to training the set and transform the data into a lower dimensional
```

```python
import matplotlib.pyplot as plot
import numpy as num
#to plot the cumulative explained variance ratio for each principal component
plot.plot(num.cumsum(PCA.explained_variance_ratio_))
#set the label for x,y axis
plot.xlabel('Components')
plot.ylabel('Variance')
```

```python
from sklearn.decomposition import PCA

PCA = PCA(n_components=54) # specify the number of principal components we want
PCA.fit(scaled_data) # fit PCA on training set features
x_PCA = PCA.transform(scaled_data) # transform training set features into lower-dimensional space of principal components
x_test_PCA = PCA.transform(X_test_scaled) # transform testing set features into same lower-dimensional space of principal components

print(scaled_data.shape) # print shape of original training set features
print(x_PCA.shape) # print shape of transformed training set features
```

```python
from sklearn import metrics#to evaluate the performance of classification model
import matplotlib.pyplot as plot

ConfusionMatrix_display = metrics.ConfusionMatrixDisplay(confusion_matrix=ConfusionMatrix, display_labels=[False, True])

# set the color of the plot to Greens
ConfusionMatrix_display.plot(cmap='Greens')
#to display the confusion matrix plot.
plot.show()
```

```python
import seaborn as sb
import numpy as num

# plot the heatmap with the color bar
ax = sb.heatmap(ConfusionMatrix/num.sum(ConfusionMatrix), annot=True, fmt='.1%', cmap='Reds', cbar=True)

# set the labels for x-axis and y-axis
ax.set_xlabel('Predicted Labels')
ax.set_ylabel('True Labels')

# set the title of the plot
ax.set_title('Confusion Matrix Heatmap')

# display the plot
plot.show()
```
